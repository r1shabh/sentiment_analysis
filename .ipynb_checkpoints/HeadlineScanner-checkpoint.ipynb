```python
from bs4 import BeautifulSoup
import pandas
from urllib.request import urlopen
import json
```

```python
def get_headlines(url):
    articles = []
    webpage = urlopen(url)
    xml_page = webpage.read()
    soup = BeautifulSoup(xml_page, 'xml')
    news_list = soup.findAll("item")
    for headline in news_list:
        article = {}
        headline_text = headline.title.text.split(" - ")
        article["Headline"] =  headline_text[0]
        article["Source"] = headline_text[1].strip()
        article["Date"] = headline.pubDate.text
        article["Sentiment"] = ''
        articles.append(article)
    return articles
```

```python
def write_to_json(content, filename):
    with open(filename, 'w') as outfile:
        json.dump(content, outfile)

def read_from_json(filename):
    with open(filename) as infile:
        content = json.load(infile)
    return content
```

```python
url = "https://news.google.com/news/rss"
articles = get_headlines(url)
write_to_json(articles, 'articles.json')
```
